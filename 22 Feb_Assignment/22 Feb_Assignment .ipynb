* DataFrame is a two-dimensional labeled data structure that is capable of holding data in various formats such as a list, dictionary, ndarray, or another DataFrame. It is similar to a spreadsheet or SQL table, with columns (variables) and rows (observations).

```python
import pandas as pd
```

```python
data1 ={'Name':['md','golam','sarwar'],'Age':[25,27,29]}
```

* Series is a one-dimensional labeled data structure that can hold any data type such as integers, strings, floats, Python objects, etc. It is similar to a column in a spreadsheet or a single column in a database table.

```python
import pandas as pd
```

```python
list = [11,22,33,44,55,66,77,88,99,100]
```

```python
import pandas as pd

fruits = pd.Series(['apple', 'banana', 'cherry', 'date'])
quantities = pd.Series([20, 25, 50, 60])
prices = pd.Series([100,60,150,600])

df = pd.DataFrame({
    'Fruit': fruits,
    'Quantity': quantities,
    'Price': prices
})

print(df)
```
